```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Example synthetic data (replace with your project dataset)
np.random.seed(0)
n = 100
df = pd.DataFrame({
    'income': np.random.normal(60000, 15000, n).astype(int),
    'monthly_spend': np.random.normal(2000, 600, n).astype(int),
    'credit_score': np.random.normal(680, 50, n).astype(int)
})
df.head()
```

## TODO: Implement at least 2 engineered features here

```python
df['spend_income_ratio'] = df['monthly_spend'] / df['income']
df['spend_income_ratio'].describe()

plt.figure(figsize=(6,4))
sns.histplot(df['spend_income_ratio'].dropna(), bins=30, kde=True)
plt.title("Distribution of Spend-to-Income Ratio")
plt.xlabel("Spend-to-Income Ratio")
plt.ylabel("Count")
plt.show()

corrs = df[['spend_income_ratio', 'monthly_spend', 'income']].corr()
print(corrs)
sns.heatmap(corrs, annot=True, center=0)
plt.title("Correlation: Ratio vs Monthly Spend & Income")
plt.show()
```

### Rationale for Feature 1
This feature captures the proportionality of spending relative to earning. While raw income and spending levels provide absolute values, the ratio helps normalize across individuals or households, making comparisons more meaningful. The histogram of the ratio shows whether most individuals spend within a narrow band of their income or if some spend disproportionately more. The correlation table shows how much new information it adds relative to `monthly_spend` and `income`. A weak correlation with income but stronger correlation with spend suggests the ratio introduces distinct explanatory power, which may improve downstream modeling.  

```python
df['credit_score_category'] = pd.cut(df['credit_score'], bins=[0, 579, 669, 739, 799, 850], labels=['Poor', 'Fair', 'Good', 'Very Good', 'Excellent'])
print(df['credit_score_category'].value_counts())

plt.figure(figsize=(6,4))
sns.countplot(x='credit_score_category', data=df,
              order=['Poor','Fair','Good','Very Good','Excellent'])
plt.title("Distribution of Credit Score Categories")
plt.xlabel("Credit Score Category")
plt.ylabel("Count")
plt.show()
```

### Rationale for Feature 2
To make credit scores more interpretable, I transformed the continuous credit score variable into categorical risk tiers following widely recognized FICO ranges. This grouping helps distinguish different consumer profiles, since small differences in raw scores may not carry much meaning, but moving between categories can indicate significant changes in creditworthiness and financial behavior. The countplot shows how observations are distributed across categories. By converting credit score into a categorical feature, models can better capture nonlinear effects and relationships that may be lost when treating credit score as a purely continuous variable.
